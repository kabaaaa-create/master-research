```python
import torch
from diffusers import MotionAdapter, AnimateDiffPipeline, DDIMScheduler
from diffusers.utils import export_to_gif

# Load the motion adapter
adapter = MotionAdapter.from_pretrained("guoyww/animatediff-motion-adapter-v1-5-2")
# load SD 1.5 based finetuned model
model_id = "Jccqqqqq/randomemotion"
pipe = AnimateDiffPipeline.from_pretrained(model_id, motion_adapter=adapter).to("cuda")

# from diffusers import DiffusionPipeline

# pipeline = DiffusionPipeline.from_pretrained("Jccqqqqq/randomemotion",safety_checker=None, torch_dtype=torch.float16)
# pipeline.to("cuda")

```

```python
import gradio as gr
import random
from transformers import pipeline
count = 1
classifier = pipeline("audio-classification", model="ehcalabres/wav2vec2-lg-xlsr-en-speech-emotion-recognition")

def classify(audio):
  emotions = classifier(audio)
  values = []
  for emotion in emotions:
    values.append(emotion["score"])
  max_value = max(values) # get max value
  max_index = values.index(max_value) # get its index
  print(emotions[max_index]["label"])
  value = values[max_index]
  print(value)
  list1 = values
  emo = emotions[max_index]["label"]
  mx = max(list1[0], list1[1])
  secondmax = min(list1[0], list1[1])
  n = len(list1)
  for i in range(2,n):
    if list1[i] > mx:
      secondmax = mx
      mx = list1[i]
    elif list1[i] > secondmax and mx != list1[i]:
      secondmax = list1[i]
    elif mx == secondmax and secondmax != list1[i]:
      secondmax = list1[i]
  max_index2 = values.index(secondmax)
  emo2 = emotions[max_index2]["label"]
  print(emo2)
  if(emo == "angry"):
    pipe.set_adapters(["pan-left","rolling-anticlockwise"], adapter_weights=[1-value, 1-value + 0.02])
    emo = "anger"
  if(emo == "disgust"):
    pipe.set_adapters(["rolling-anticlockwise","pan-right"], adapter_weights=[1-value, 1-value + 0.02])
  if(emo == "fearful"):
    pipe.set_adapters(["rolling-anticlockwise","pan-right"], adapter_weights=[1-value, 1-value + 0.02])
    emo = "fear"
  if(emo == "sad"):
    pipe.set_adapters(["rolling-anticlockwise","pan-left"], adapter_weights=[1-value, 1-value + 0.02])
  if(emo == "happy"):
    pipe.set_adapters(["rolling-clockwise","zoom-out"], adapter_weights=[1-value, 1-value + 0.02])
  if(emo == "surprised"):
    pipe.set_adapters(["rolling-clockwise","zoom-out"], adapter_weights=[1-value, 1-value + 0.02])
    emo = "surprise"
  if(emo == "calm"):
    emo = "happy"
    pipe.set_adapters(["rolling-clockwise","zoom-out"], adapter_weights=[1-value, 1-value + 0.02])
  if(emo == "neutral"):
    emo = "happy"
    pipe.set_adapters(["rolling-clockwise","zoom-out"], adapter_weights=[1-value, 1-value + 0.02])
  print(emo)
  image = pipeline2(emo + " " + emo2, num_inference_steps=50).images[0]
  output = pipe(prompt=("{}".format(emo + " " + emo2)),num_frames=16,guidance_scale=7.5,num_inference_steps=25,ip_adapter_image=image,generator=torch.Generator("cuda").manual_seed(random.randrange(999)),)
  frames = output.frames[0]
  export_to_gif(frames, "animation-{}-{}.gif".format(count,emo + " " + emo2))
  rs = "/content/animation-{}-{}.gif".format(count,emo + " " + emo2)
  return rs
demo = gr.Interface(fn=classify, inputs=gr.Audio(type="filepath"), outputs=gr.Image())
demo.queue()
demo.launch(debug=True, inline=False)
```
